```python
# find the longest movie directed by Steven Spielberg
# 1) filter then find rows equal to the subset max (handles ties)
subset = movie[movie['director'] == 'Steven Spielberg']
longest = subset[subset['duration'] == subset['duration'].max()]
print(longest)
```

```python
# 2) use idxmax to get a single row (first max)
idx = movie.loc[movie['director'] == 'Steven Spielberg', 'duration'].idxmax()
longest_row = movie.loc[idx]
print(longest_row)
```

```python
ss = movie[movie['director'] == 'Steven Spielberg']
ss[ss['duration'] == ss['duration'].max()]
```

```python
# 1. find the average temperature and maximum 
print(weather['temp'].mean())
print(weather['temp'].max())


# 2. what was the temperature on Monday and Friday?
weather[weather['day'].isin(['Monday', 'Friday'])]

# 3. What was the temperature on Saturday in Farenheit?
weather['temp_f'] = weather['temp'] * 9/5 + 32
weather[['day', 'temp_f']][weather['day'] == 'Saturday']
```
